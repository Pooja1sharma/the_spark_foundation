### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Ploting the dataset

```python
df = pd.read_csv('SampleSuperstore.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Ship Mode'].value_counts()
```

```python
plt.figure(figsize=(30,30))
df.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title('Total Proft and Sales per Sub-Category')
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.countplot(x='Sub-Category', hue='Region', data=df)
plt.show()
```

```python
plt.figure(figsize=(12,10))
df['Sub-Category'].value_counts().plot.pie(autopct="%.2f")
```

```python
sns.catplot(y="Category", hue="Region", kind="count", data=df)
```

```python
sns.scatterplot("Sales","Profit", data=df)
```

```python
pd.crosstab(df["Quantity"],df["Category"],df["Profit"],aggfunc='sum').plot(kind="bar",stacked=True)
```
